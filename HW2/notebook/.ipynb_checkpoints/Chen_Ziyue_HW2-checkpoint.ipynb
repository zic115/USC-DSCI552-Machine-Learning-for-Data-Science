### (b) iii. What are the mean, the median, range, first and third quartiles, and interquartile ranges of each of the variables in the dataset?

```python
def Stats(df, variable):
    mean = df[variable].mean()
    median = df[variable].median()
    rg = max(df[variable]) - min(df[variable])
    q1 = df[variable].quantile(0.25)
    q3 = df[variable].quantile(0.75)
    iqr = q3 - q1
    
    data = {'Mean': [mean],
            'Median': [median],
            'Range': [rg],
            'Q1': [q1],
            'Q3': [q3],
            'IQR': [iqr]}
    df = pd.DataFrame(data).rename(index = {0: variable})
    return df
```

```python
AT = Stats(df, 'AT')
V = Stats(df, 'V')
AP = Stats(df, 'AP')
RH = Stats(df, 'RH')
PE = Stats(df, 'PE')

stats_df = pd.concat([AT, V, AP, RH, PE])
stats_df
```

## (c) Fit a simple linear regression model for each predictor

```python
# PE vs. AT simple linear regression model
X = df['AT']
X = sm.add_constant(X)
Y = df['PE']

AT_model = sm.OLS(Y, X).fit()
print(AT_model.summary())
```

Observations:

1) Since the p-value is rounded to 0.000, which is smaller than 0.05, there is a statistically significant association between the predictor AT and the response PE in this simple linear regression model.

2) The coefficient is -2.1713, and the intercept is 497.0341. This gives the model $\hat{y} = 497.0341 - 2.1713x + \epsilon$.

```python
# PE vs. AT regression plot
sns.regplot(x = 'AT', y = 'PE', data = df, scatter_kws={"color": 'blue'}, line_kws={"color": 'red'})

plt.title('PE vs. AT')
plt.show()
```

Observations:

1) This regression plot indicates there is a negative correlation between PE and AT, backing up the assertion mentioned above that there is a statistically significant association betweem the predictor and the outcome.

```python
# PE vs. AT detect outliers using Cook's Distance
AT_cooks = AT_model.get_influence().cooks_distance[0]

threshold = row / 4
AT_outliers = 0
for pt in AT_cooks:
    if pt > threshold:
        AT_outliers += 1

print('There are', AT_outliers, 'outlier(s) in this linear regression model.')
```

## (d) Fit a multiple regression model using all of the predictors

```python
# PE vs. all predictors multiple regression model
X = df[['AT', 'V', 'AP', 'RH']]
X = sm.add_constant(X)
Y = df['PE']

multi_model = sm.OLS(Y, X).fit()
print(multi_model.summary())
```

Observations:

1) Since all the p-values are rounded to 0.000, which is smaller than 0.05, we can reject the null hypothesis for all predictors. It means there a statistically significant association between all the predictors AT, V, AP, and RH and the response PE in this multiple regression model.

2) This gives the model $\hat{y} = 454.6093 -1.9775x_{1} - 0.2339x_{2} + 0.0621x_{3} - 0.1581x_{4} + \epsilon$ in which $x_{1}$ is AT, $x_{2}$ is V, $x_{3}$ is AP, and $x_{4}$ is RH.

## (e) Plot displaying the univariate regression coefficients from (c) and multiple regression coefficients from (d)

```python
# Create a dataframe to record the coefficients
coef_data = [[-2.1713, -1.9775],
             [-1.1681, -0.2339],
             [1.4899, 0.0621],
             [0.4557, -0.1581]]

coef_df = pd.DataFrame(data = coef_data, columns = ['univariate regression coefficient', 'multiple regression coefficient']).rename(index = {0: 'AT', 1: 'V', 2: 'AP', 3: 'RH'})
coef_df
```

 ## (f) Nonlinear association between any of the predictors and the response?

```python
# PE vs. AT polynomial regression model
X = df['AT'].to_numpy()
Y = df['PE']

poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X.reshape(-1, 1))

AT_poly_model = sm.OLS(Y, X_poly).fit()
print(AT_poly_model.summary())
```

Observations:

1) Since all the p-values are rounded to 0.000, which is smaller than 0.05, there a statistically significant association between the predictor AT and the response PE in this polynomial regression model.

2) This gives the model $\hat{y} = 492.7281 - 0.6103x - 0.1251x^2 + 0.0027x^3 + \epsilon$.

```python
# PE vs. V polynomial regression model
X = df['V'].to_numpy()
Y = df['PE']

poly = PolynomialFeatures(degree = 3)
X_poly = poly.fit_transform(X.reshape(-1, 1))

V_poly_model = sm.OLS(Y, X_poly).fit()
print(V_poly_model.summary())
```

Observations:

1) Since the p-value for x2 is 0.768, which is greater than 0.05, there is a statistically insignificant association between the predictor V^2 and the response PE in this polynomial regression model. However, based on the hierarchical principle, the V^2 term may be retained as the higher order V^3 term is statistically significant by having a p-value smaller than 0.05.

2) This gives the model $\hat{y} = 554.1468 - 2.1444x - 0.0027x^2 + 0.0001x^3 + \epsilon$.

## (g) Association of interactions of predictors?

```python
# PE vs. (all predictors + pairwise interactions) full linear regression model
X = df[['AT', 'V', 'AP', 'RH']]
Y = df['PE']

poly = PolynomialFeatures(interaction_only = True)
X_poly = poly.fit_transform(X)

full_model = sm.OLS(Y, X_poly).fit()
print(full_model.summary())
```

```python
# Match x with predictors and pairwise interactions
print(poly.get_feature_names_out())
```

Observations:

1) Since the p-values for 'AT V', 'AT RH', 'V AP', and 'AP RH' are smaller than 0.05, there is a statistically significant association between the interaction terms 'AT V', 'AT RH', 'V AP', and 'AP RH' and the response PE in this full linear regression model.

2) Since the p-values for 'AT AP' and 'V RH' are greater than 0.05, there is a statistically insignificant association between the interaction terms 'AT AP' and 'V RH' and the response PE in this full linear regression model.

## (h) Improve the model

```python
# Split the data
train, test = train_test_split(df, train_size = 0.7, random_state = 42)
```

```python
# PE vs. all predictors multiple regression model with training size = 0.7
X_train = train[['AT', 'V', 'AP', 'RH']]
X_train = sm.add_constant(X_train)
Y_train = train['PE']

train_multi_model = sm.OLS(Y_train, X_train).fit()
print(train_multi_model.summary())
```

```python
# Test train_multi_model on train and test data
X_test = test[['AT', 'V', 'AP', 'RH']]
X_test = sm.add_constant(X_test)
Y_test = test['PE']

y_train_pred = train_multi_model.predict(X_train)
tr_mse = mean_squared_error(Y_train, y_train_pred)

y_test_pred = train_multi_model.predict(X_test)
te_mse = mean_squared_error(Y_test, y_test_pred)
```

```python
# Record train and test MSEs
data = {'Train MSE': [tr_mse],
        'Test MSE': [te_mse]}
multi_model_stats = pd.DataFrame(data).rename(index = {0: 'All predictors'})
multi_model_stats
```

```python
# PE vs. (all predictors + pairwise interactions + quadratic nonlinearities) regression model with training size = 0.7
X_train = train[['AT', 'V', 'AP', 'RH']]
Y_train = train['PE']

poly = PolynomialFeatures()
X_poly = poly.fit_transform(X_train)

train_all_model = sm.OLS(Y_train, X_poly).fit()
print(train_all_model.summary())
```

```python
# Match x with predictors, pairwise interactions, and quadratic nonlinearities
print(poly.get_feature_names_out())
```

Insignificant terms (p > 0.05): x2 ('V'), x7 ('AT AP'), x9 ('V^2'), x10 ('V AP'), x11 ('V RH')

Significant terms (p <= 0.05): '1', 'AT', 'AP', 'RH', 'AT^2', 'AT V', 'AT RH', 'AP^2', 'AP RH', 'RH^2'

Based on the hierarchical principle:

1) 'V' should be retained because 'AT V' is significant.

2) 'AT AP', 'V^2', 'V AP', and 'V RH' can be dropped.

```python
# Drop insignificant terms and fit again
X_train = X_poly[:, [0, 1, 2, 3, 4, 5, 6, 8, 12, 13, 14]]
Y_train = train['PE']

train_all_sig_model = sm.OLS(Y_train, X_train).fit()
print(train_all_sig_model.summary())
```

```python
# Test train_multi_model on train and test data
X_test = test[['AT', 'V', 'AP', 'RH']]
Y_test = test['PE']

poly = PolynomialFeatures()
X_poly = poly.fit_transform(X_test)
X_test = X_poly[:, [0, 1, 2, 3, 4, 5, 6, 8, 12, 13, 14]]

y_train_pred = train_all_sig_model.predict(X_train)
tr_mse = mean_squared_error(Y_train, y_train_pred)

y_test_pred = train_all_sig_model.predict(X_test)
te_mse = mean_squared_error(Y_test, y_test_pred)
```

```python
# Record train and test MSEs
data = {'Train MSE': [tr_mse],
        'Test MSE': [te_mse]}
all_sig_model_stats = pd.DataFrame(data).rename(index = {0: 'All significant terms'})
all_sig_model_stats
```

```python
# Combine the MSEs for two models
combined = pd.concat([multi_model_stats, all_sig_model_stats])
combined
```
